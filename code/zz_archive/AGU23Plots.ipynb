```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as pltc
import matplotlib as mpl
import geopandas
import glob 
from sklearn import preprocessing

mpl.rcParams['pdf.fonttype'] = 42

## Cubic feet to cubic meters conversion factor
cfs_2_cms = 0.0283168466
```

```python
pnwNP = pd.read_csv("../../data/pnwNP_Info.csv")
# pnwNP
```

```python

plt.rcParams.update({'font.size': 18})

gages = [14362250,12115700,14034470]
for i in gages:
    print(i)

    datMain = pd.read_csv('../../data/pnwNP_modeledData/'+str(i)+".csv").drop(['Unnamed: 0','gage'],axis=1)
    datMain['time'] = pd.to_datetime(datMain['time'])
    datMain['time'] = datMain['time'].dt.tz_localize(None)


    ######################
    ## Plot streamflow from 2000 - 2004
    ######################

    sub = datMain[ (datMain["time"] >= '2000-01-01') & (datMain["time"] <= '2004-12-31')]
    ax = sub.plot(x = "time",figsize = (25,10),logy=True,alpha=0.6,color = ['purple','green','crimson','orangered','grey'],legend=False)
    ax.set_xlabel('Time')
    ax.set_ylabel('Streamflow $(m^3/s)$')

    plt.savefig('../../docs/pnwModeled/'+str([i])+"_streamflowSubset.pdf")
#     plt.savefig('../docs/pnwModeled/'+str(shp['gage'][i])+"_streamflowSubset.png",dpi=300)
```
